```python
import tensorflow as tf
import numpy as np
```

```python
# Define input data
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
# Define output data
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)
```

This model defines a simple neural network with 1 layer and 1 neuron.

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
```

หลักการทำงาน ของ SGD ทำงานโดย:

คำนวณ Gradients:

เริ่มต้นด้วยโมเดลที่มีพารามิเตอร์สุ่ม
นำข้อมูลชุดย่อย (mini-batch) มาป้อนโมเดลexpand_more
คำนวณค่าความสูญเสีย (loss) ของโมเดลสำหรับข้อมูลชุดย่อยนั้น
คำนวณ Gradients ของค่า loss เทียบกับพารามิเตอร์ของโมเดล
ปรับแต่งพารามิเตอร์:

อัปเดตค่าพารามิเตอร์ของโมเดลในทิศทางตรงกันข้ามกับ Gradients
อัตราการปรับเปลี่ยนจะถูกควบคุมโดย Learning Rate
ทำซ้ำ:

ทำขั้นตอน 1 และ 2 ซ้ำๆ กับข้อมูลชุดย่อยอื่นๆ จนกว่าจะบรรลุเกณฑ์การหยุด
ข้อดี ของ SGD:

มีประสิทธิภาพในการคำนวณสูง เหมาะสำหรับโมเดลขนาดใหญ่
เรียบง่าย เข้าใจง่าย
ปรับแต่งได้หลากหลาย
ข้อเสีย ของ SGD:

อาจติดอยู่ใน Local Minima ได้ง่าย
อาจมีความผันผวนสูงในช่วงต้นของการฝึกอบรม
อาจต้องใช้ Learning Rate ที่ต่ำ ซึ่งทำให้การฝึกอบรมช้าลง

MeanSquaredError MSE = 1/n * Σ(y_true - y_pred)^2
ยิ่งน้อยแสดงว่าเราเข้าใกล้คำตอบแล้วไงล่ะหนุ่ม แต่ถ้า 0 คือตรงกับคำตอบเลย
ข้อดีของ MSE:

เข้าใจง่าย คำนวณง่าย
มีความไวต่อ outliers น้อย
เหมาะสำหรับข้อมูลที่เป็น continuous data
ข้อเสียของ MSE:

ไม่สามารถวัดความผิดพลาดเชิงคุณภาพ (qualitative error) ได้
ไม่เหมาะสำหรับข้อมูลที่เป็น categorical data
อาจมีปัญหาเมื่อค่าที่แท้จริงมีค่าใกล้เคียงกับ 0

```python
model.compile(optimizer=tf.keras.optimizers.SGD(), loss=tf.keras.losses.MeanSquaredError())
```

ได้เวลาเทรนแล้ว ในที่นี้เราจะใช้ 500 epochs นะครับ คือโมเดลเราจะถูกสอนโดยชุดข้อมูลที่เราเตรียมไว้ 500 ครั้ง จริงๆแล้วโมเดลไม่ซับซ้อนแบบนี้ใช้ 50 ครั้ง google บอกส่าน่าจะพอละ ดังนั้นถ้าไม่อยากให้คอมทำงานหนักลดได้นะ

```python
model.fit(xs, ys, epochs=500)
```

จะสังเกตว่าระหว่างรันค่า loss จะลดลงเรื่อยๆ แสดงว่าโมเดลเราเริ่มเรียนรู้ได้ดีขึ้น และค่า loss จะลดลงจนเข้าใกล้ 0 หรือเข้าใกล้คำตอบที่ถูกต้องมากที่สุด แต่ก็อย่าคิดว่าจะดีเสมอไปเพราะมันอาจจะไปติดกับการ Local Minimal

```python
result = model.predict([10.0])
print(result)
```
